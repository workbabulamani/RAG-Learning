# Word document processing

```python
from docx import Document as DocxDocument
import os
```

```python
from langchain_community.document_loaders import Docx2txtLoader,UnstructuredWordDocumentLoader
```

# Method 1: Using Docx3txtLoader

```python
try:
    docx_loader = Docx2txtLoader("data/docs/paper.docx")
    docs = docx_loader.load()
    print(f"Loaded {len(docs)} document(s)")
    print(f"COntent preview: {docs[0].page_content[:100]}...")
    print(f"Metadata: {docs[0].metadata}")
except Exception as e:
    raise Exception(str(e))
```

## Method 2: UnstructuredWordDocumentLoader

```python
try:
    unstructured_loader = UnstructuredWordDocumentLoader("data/docs/paper.docx",
                                                 mode="elements") # mode is to follow structure of the document
    unstructured = unstructured_loader.load()
    print(f"Loaded {len(unstructured)} document(s)")
    print(f"Content preview: {unstructured[0].page_content[:100]}...")
    print(f"Metadata: {unstructured[0].metadata}")
except Exception as e:
    raise Exception(str(e))
```
